```python
import numpy as np

def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calculate great-circle distance between two points on the Earth (Haversine formula).
    Inputs are in decimal degrees.
    Returns distance in kilometers.
    """
    R = 6371  # Earth radius in kilometers
    
    # Convert decimal degrees to radians
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    # Haversine formula
    a = np.sin(delta_phi / 2.0)**2 + \
        np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    distance = R * c
    return distance
```

```python
merged_df['distance_km'] = haversine_distance(
    merged_df['pickup_latitude'],
    merged_df['pickup_longitude'],
    merged_df['dropoff_latitude'],
    merged_df['dropoff_longitude'])

merged_df.head()
```

```python
Q1 = merged_df['trip_duration(min)'].quantile(0.25)
Q3 = merged_df['trip_duration(min)'].quantile(0.75)
IQR = Q3 - Q1
```
